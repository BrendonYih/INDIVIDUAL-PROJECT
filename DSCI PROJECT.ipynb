# 2. Broad Question

Can a player's playing hours, subscription status, and demographic factors predict their experience level?

## Specific Question
Does a player’s playing hours and age predict their experience level, and how does experience level contribute to the overall data? 

### Response Variable (Dependent Variable):  
- experience (Categorical: Pro, Veteran, Regular, Amateur, Beginner)

### Explanatory Variables (Independent Variables):  
- played_hours (Numerical)  
- Age (Numerical)  

The project answers the broad question of how different kinds of players are most likely to contribute a large amount of data. Players with more experience tend to involve themselves more in the game, so their playing hours and age should yield a large amount of data concerning the game’s analytics, gaming styles, etc. Since we are assuming that experienced players are the ones who contribute the most data, we must be able to predict experience from other variables. The project will answer the specific question of whether or not we are able to predict a player’s experience from two numerical variables, played hours and age, through K-nearest classification. Computing summary statistics for numerical variables and using the K-nearest neighbour will establish the closest prediction for the experience with respect to a player’s age and number of played hours. Since experience is a categorical variable, we must first factor it to ensure that it is properly encoded. We will need to set up our own k model, which will be used to help us form our prediction. We will need to handle missing values using the filter function with na.rm = TRUE to remove anything that may cause potential errors or biases in the dataset. To make sure everything is properly scaled (mean = 0 and our standard deviation = 1), we can use our step.scale functions. We will visualize the results using a scatter plot, which is the best for performing K-nearest analysis. We will explore other alternatives if these methods do not work.

# 3. Analysis Plan






```r
# Load libraries
library(tidyverse)
library(ggplot2)
library(janitor)
library(readr)
library(kknn)

# Load dataset
players <- read.csv("https://raw.githubusercontent.com/BrendonYih/INDIVIDUAL-PROJECT/refs/heads/main/players.csv")
players

sessions <- read.csv("https://raw.githubusercontent.com/BrendonYih/INDIVIDUAL-PROJECT/refs/heads/main/sessions.csv")
sessions

# Convert categorical variables to factors and numeric variables to numeric
players <- players |> 
  mutate(
    experience = as.factor(experience),
    subscribe = as.factor(subscribe),
    gender = as.factor(gender),
    played_hours = as.numeric(played_hours),  
    Age = as.numeric(Age)                    
  ) |> 
  drop_na()  # Remove rows with missing values

# Compute mean values
mean_values <- players |> 
  summarize(
    mean_played_hours = mean(played_hours, na.rm = TRUE),
    mean_age = mean(Age, na.rm = TRUE)
  )

# Print as table
mean_values |> knitr::kable()

# Histogram for played_hours
ggplot(players, aes(x = log1p(played_hours))) +
  geom_histogram(binwidth = 0.3, fill = "blue", color = "black", alpha = 0.7) +
  labs(title = "Log-Transformed Distribution of Played Hours", 
       x = "Played Hours", 
       y = "Count") +
  theme_minimal()

# Histogram for Age
ggplot(players, aes(x = Age)) +
  geom_histogram(binwidth = 2, fill = "darkgreen", color = "black", alpha = 0.7) +
  labs(title = "Distribution of Player Ages", 
       x = "Age (Years)", 
       y = "Count") +
  theme_minimal()
```

# 4. Approach: K-Nearest Neighbors (KNN) Classification

The chosen method for this analysis is K-Nearest Neighbors (KNN) classification. It is well-suited for predicting a categorical variable—experience level—based on numerical features such as played hours and age. KNN is a learning algorithm that assigns a class label based on the majority class of the k-nearest data points. This method is appropriate because it makes minimal assumptions about the data distribution, allowing for flexibility in capturing relationships between the explanatory variables and the response variable. However, KNN assumes that the data is properly scaled, and it can be sensitive to irrelevant or highly mixed features. Additionally, it may struggle with imbalanced class distributions and high-dimensional data. Thus, we will ensure proper feature scaling using standardization (mean = 0, standard deviation = 1) and test different values of k to find the optimal number of neighbors. To choose the best version of our KNN model, we will test different values of k (the number of neighbors) and compare their accuracy. Instead of just picking one k-value randomly, we will check multiple k-values and see which one gives the best performance. We will also use cross-validation by breaking the training set into smaller parts and testing on different sections.This helps avoid overfitting and ensures the model works well on different data. For example, we can split the players into two distinct date ranges from the sessions dataset to compare older and newer data.
